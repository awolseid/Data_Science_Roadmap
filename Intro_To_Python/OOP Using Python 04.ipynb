```python
from enum import Enum
from functools import total_ordering
from random import shuffle

class Course(Enum):
  AI = "Artificial Intelligence"
  ML = "Machine Learning"
  # DL = "Deep Learning" # not in the data
  Stat = "Statistics"
  Maths = "Mathematics"
  Prog = "Programming"
  # Econ = "Economics" # not in the data

class Rank(Enum):
  Prof = "Professor"
  AssocProf = "Associate Professor"
  AssisProf = "Assistant Professor"
  Lecturer = "Lecturer"
  GraduateAssistant = "Graduate Assistant"

class Instructor:
  all_instructors = []
  def __init__(self, first_name, last_ame, rank: Rank):
    self.first_name = first_name
    self.last_ame = last_ame
    self.rank = rank
    self.courses = []
    Instructor.all_instructors.append(self)

  def __repr__(self):
    return f"{self.first_name} {self.last_ame} - {self.rank}"
  @property
  def rank(self):
    return self._rank.value # change is here
  @rank.setter
  def rank(self, new_rank):
    if isinstance(new_rank, Rank):
      new_rank = new_rank.name
    if new_rank in Rank.__members__:
      self._rank = Rank[new_rank]
    else:
      raise ValueError(f"Rank = {new_rank} is not in the rank list.")

  def assign_course(self, *subject):
    for course in subject:
      if isinstance(course, Course):
        course = course.name
      if course not in Course.__members__:
        print(course)
        raise ValueError(f"{course} is not offered.")
      if Course[course].value not in self.courses:
        self.courses.append(Course[course].value)
        if Department.course_instructor.get(course) is None:
          Department.course_instructor[course] = [self]
        else:
          Department.course_instructor[course].append(self)



class Student:
  dept = "Data Science"
  all_students = []
  def __init__(self, f_name, l_name, r_num):
    self.first_name = f_name
    self.last_name = l_name
    self.reg_num = r_num
    self.email = f"{self.first_name.lower()}.{self.last_name.lower()}@pukyong.ac.kr"
    self.courses = [Course.Stat.value]
    self.group_mem = None
    Student.all_students.append(self)

  @property
  def full_name(self):
    return f"{self.first_name} {self.last_name}"

  @full_name.setter
  def full_name(self, new_full_name):
    f, l = new_full_name.split(" ")
    self.first_name = f
    self.last_name = l

  @property
  def first_name(self):
    return self._first_name

  @first_name.setter
  def first_name(self, new_name):
    if Student.valid_name(new_name):
      self._first_name = new_name
    else:
      raise ValueError("Name must be alphabets only, with atleast 2 characters.")

  @property
  def last_name(self):
    return self._last_name

  @last_name.setter
  def last_name(self, new_name):
    if Student.valid_name(new_name):
      self._last_name = new_name
    else:
      raise ValueError("Name must be alphabets only, with atleast 2 characters.")

  @property
  def reg_num(self):
    return self._reg_num

  @reg_num.setter
  def reg_num(self, new_reg_num):
    if isinstance(new_reg_num, str) and str(new_reg_num).startswith("UG-"):
      self._reg_num = new_reg_num
    else:
      raise ValueError("Registration number does not start with 'UG-'")

  @staticmethod
  def valid_name(new_name):
    if (isinstance(new_name, str) and len(new_name) >= 2 and new_name.isalpha()):
      return True
    else:
      return False

  def register_course(self, *subject):
    for course in subject:
      if isinstance(course, Course):
        course = course.name
      if course not in Course.__members__:
        print(course)
        raise ValueError(f"{course} is not offered.")
      if Course[course].value not in self.courses:
        self.courses.append(Course[course].value)
        if Department.course_student.get(course) is None:
          Department.course_student[course] = [self]
        else:
          Department.course_student[course].append(self)

  @classmethod
  def not_reg_courses(cls): #cls is a convension
      a = set()
      m = [course.value for course in Course]
      for stud in cls.all_students:
        s = set(stud.courses)
        a.update(s)
      return list(set(m).difference(a))

  @classmethod
  def without_group_mem(cls):
    return list(filter(lambda s: s.group_mem is None, cls.all_students))

  ## Magic Methods ##
  def __repr__(self):
    return f'{self.last_name}-{self.reg_num[-2: ]}'
  def __add__(self, other):
    if self.group_mem is not None:
      raise ValueError(f"{self} already has {self.group_mem} group.")
    elif other.group_mem is not None:
      raise ValueError(f"{other} already has {other.group_mem} group.")
    else:
      self.group_mem = other
      other.group_mem = self

  def __sub__(self, other):
    if self.group_mem is None and other.group_mem is None:
      return
    elif self.group_mem != other:
      raise ValueError(f"{self} is not a group member of {other}.")
    else:
      self.group_mem = None
      other.group_mem = None

  def __lt__(self, other):
    # to compare number of courses
    return len(self) < len(other)
  def __eq__(self, other):
    # to compare number of courses
    return len(self) == len(other)

  def __len__(self):
    return len(self.courses)

  def __iter__(self):
    yield self.first_name
    yield self.last_name
    yield self.reg_num
    for c in self.courses:
      yield c

class Department:
  course_instructor = {}
  course_student = {"Stat": Student.all_students}
  @staticmethod
  def create_groups():
    shuffle(Student.all_students)
    l = len(Student.all_students)
    for s in range(0, l, 2):
      if s != l-1:
        Student.all_students[s] + Student.all_students[s+1]
        #if odd number of students, last student will not be assigned to any group
        # we left it here as it is., Instructor will assign that student whichever group s/he wants
```
